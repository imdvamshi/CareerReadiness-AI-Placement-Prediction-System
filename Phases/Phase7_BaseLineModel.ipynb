```python
import pandas as pd
import numpy as np
```

```python
import joblib
```

```python
x_train = pd.read_csv("../src/x_train.csv")
y_train = pd.read_csv("../src/y_train.csv")
y_train = y_train.squeeze()
x_test = pd.read_csv("../src/x_test.csv")
y_test = pd.read_csv("../src/y_test.csv")
y_test = y_test.squeeze()
```

```python
x_train.drop(columns=['student_id'],inplace=True)
x_test.drop(columns=['student_id'],inplace=True)
```

ColumnTransformer 

```python
# numeric columns 
num_cols = [
    "age","cgpa","internships_count","projects_count",
    "certifications_count","coding_skill_score",
    "aptitude_score","communication_skill_score",
    "logical_reasoning_score","hackathons_participated",
    "github_repos","linkedin_connections",
    "mock_interview_score","attendance_percentage",
    "backlogs","extracurricular_score","leadership_score",
    "sleep_hours","study_hours_per_day"
]
```

```python
# text columns for OneHotEncoder
text_cols_onehot = [
    "gender","branch","volunteer_experience"
]
```

```python
# text columns for OrdinalEncoder
text_cols_ordinal = ["college_tier"]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
preprocessor = ColumnTransformer(transformers=[
    ('ohe',OneHotEncoder(handle_unknown='ignore'),text_cols_onehot),
    ('oe',OrdinalEncoder(),text_cols_ordinal),
    ('scaler',StandardScaler(),num_cols)
])
```

### BASELINE MODEL -- TRAINING AND EVALUATION 

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
modelLR = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('lr',LogisticRegression())
])
modelLR.fit(x_train,y_train)
predictionLR = modelLR.predict(x_test)
```

```python
# result of logistic regression
from sklearn.metrics import (confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score)
cm = confusion_matrix(y_test,predictionLR)
acc_train = modelLR.score(x_train,y_train)
acc_test = modelLR.score(x_test,y_test)
p = precision_score(y_test,predictionLR)
r = recall_score(y_test,predictionLR)
f1 = f1_score(y_test,predictionLR)
roc = roc_auc_score(y_test,modelLR.predict_proba(x_test)[:,1])
print("train_accuracy score :",acc_train)
print("test_accuracy score :",acc_test)
print("precision :",p)
print("recall :",r)
print("f1 score :",f1)
print("roc auc score :",roc)
print("confusion matrix :",cm)
```

```python
joblib.dump(modelLR,'../Models/LogisticRegression_Baselinemodel.pkl')
```

Voting

```python
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline
modelV = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('v',VotingClassifier(estimators=[
        ('lr',LogisticRegression()),
        ('knn',KNeighborsClassifier()),
        ('dt',DecisionTreeClassifier()),
        ('rf',RandomForestClassifier()),
        ('svm',SVC(probability=True)),
        ('nb',GaussianNB()),
        ('xg',XGBClassifier())
    ],voting='soft'))
])
modelV.fit(x_train,y_train)
predictionV = modelV.predict(x_test)
```

```python
# result of Voting
from sklearn.metrics import (accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score)
cm = confusion_matrix(y_test,predictionV)
acc_train = accuracy_score(y_train,modelV.predict(x_train))
acc_test = accuracy_score(y_test,predictionV)
p = precision_score(y_test,predictionV)
r = recall_score(y_test,predictionV)
f1 = f1_score(y_test,predictionV)
roc = roc_auc_score(y_test, modelV.predict_proba(x_test)[:, 1])
print("train_accuracy score :",acc_train)
print("test_accuracy score :",acc_test)
print("precision :",p)
print("recall :",r)
print("f1 score :",f1)
print("roc auc score :",roc)
print("confusion matrix :",cm)
```

Stacking

```python
from sklearn.ensemble import StackingClassifier
from sklearn.pipeline import Pipeline
modelS = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('s',StackingClassifier(estimators=[
        ('lr',LogisticRegression()),
        ('knn',KNeighborsClassifier()),
        ('dt',DecisionTreeClassifier()),
        ('rf',RandomForestClassifier()),
        ('svm',SVC()),
        ('nb',GaussianNB()),
        ('xg',XGBClassifier())
    ], final_estimator=LogisticRegression()))
])
modelS.fit(x_train,y_train)
predictionS = modelS.predict(x_test)
```

```python
# result of Stacking
from sklearn.metrics import (accuracy_score,confusion_matrix,precision_score,recall_score,f1_score,roc_auc_score)
cm = confusion_matrix(y_test,predictionS)
acc_train = accuracy_score(y_train,modelS.predict(x_train))
acc_test = accuracy_score(y_test,predictionS)
p = precision_score(y_test,predictionS)
r = recall_score(y_test,predictionS)
f1 = f1_score(y_test,predictionS)
roc = roc_auc_score(y_test,modelS.predict_proba(x_test)[:,1])
print("train_accuracy score :",acc_train)
print("test_accuracy score :",acc_test)
print("precision :",p)
print("recall :",r)
print("f1 score :",f1)
print("roc auc score :",roc)
print("confusion matrix :",cm)
```
